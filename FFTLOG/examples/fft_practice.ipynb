```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_num = 100
sample_interval = 0.1
sample_rate = 1 / sample_interval # sampling rate
Nq = sample_rate/2
print('Nq=',Nq)

x= np.array([i*sample_interval for i in range(sample_num)])
freq=[1,2,3]
amp=[3,2,1]
yx = amp[0]*np.sin(freq[0]*(np.pi*2)*x) + amp[1]*np.sin(freq[1]*(np.pi*2)*x)+amp[2]*np.sin(freq[2]*(np.pi*2)*x)

plt.plot(x,yx)
plt.xlim(0,2)
plt.ylabel('y(t)')
plt.xlabel('t')
```

```python
#k = np.fft.fftfreq(x.shape[0],d=sample_interval)
k = np.fft.fftfreq(x.shape[0])*sample_rate
Yk = np.fft.fft(yx)

k_shift = np.fft.fftshift(k)
Yk_shift = np.fft.fftshift(np.abs(Yk))
plt.plot(k_shift, Yk_shift)
plt.xlim(0,Nq)

plt.ylabel('Y(k)')
plt.xlabel('k')
```
