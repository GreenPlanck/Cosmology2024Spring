```python
import os
import numpy as np

class bao_boss:
    def __init__(self):
        
        
        self.z = np.array([], 'float64')
        self.data = np.array([], 'float64')
        self.error = np.array([], 'float64')
        self.type = np.array([], 'int')

        # read redshifts and data points
        with open('/Users/zhiyulu/Documents/Github/CourseCosmology2024/Sampler/cosmo_test/bao_data.txt', 'r') as filein:
            for line in filein:
                if line.strip() and line.find('#') == -1:
                    # the first entry of the line is the identifier
                    this_line = line.split()
                    # insert into array if this id is not manually excluded
                    self.z = np.append(self.z, float(this_line[1]))
                    self.data = np.append(self.data, float(this_line[2]))
                    self.error = np.append(self.error, float(this_line[3]))
                    self.type = np.append(self.type, int(this_line[4]))

        # number of data points
        self.num_points = np.shape(self.z)[0]

        # end of initialization

    # compute likelihood

    def loglkl(self, cosmo):

        chi2 = 0.

        # for each point, compute angular distance da, radial distance dr,
        # volume distance dv, sound horizon at baryon drag rs_d,
        # theoretical prediction and chi2 contribution
        for i in range(self.num_points):

            da = cosmo.angular_distance(self.z[i])
            dr = self.z[i] / cosmo.Hubble(self.z[i])
            dv = pow(da * da * (1 + self.z[i]) * (1 + self.z[i]) * dr, 1. / 3.)
            rs = cosmo.rs_drag()  #  use this value for simplity self.rs=149.92  # [Mpc]

            if self.type[i] == 3:
                theo = dv / rs

            elif self.type[i] == 4:
                theo = dv

            elif self.type[i] == 5:
                theo = da / rs

            elif self.type[i] == 6:
                theo = 1. / cosmo.Hubble(self.z[i]) / rs

            elif self.type[i] == 7:
                theo = rs / dv
            else:
                raise ValueError('wrong type')

            chi2 += ((theo - self.data[i]) / self.error[i]) ** 2

        # return ln(L)
        lkl = - 0.5 * chi2

        return lkl
```

```python
from classy import Class
m=Class()
bao=bao_boss()
```

```python
def get_theory(theta,cosmo):
    Omega_m,Omega_b,H0=theta
    cosmo.set({'H0':H0,
                 'Omega_m':Omega_m,
                 'Omega_b':Omega_b})
    cosmo.compute()
    return cosmo
```

```python
theta=(0.25,0.05,70)
```

```python
def loglkl(theta,cosmo=None):
    cosmo=get_theory(theta,cosmo)
    return bao.loglkl(cosmo)

def log_prior(theta):
    Omega_m,Omega_b,H0=theta
    if 0.2 < Omega_m < 0.4 and 0.01 < Omega_b < 0.1 and 60 < H0 < 80 :
        return 0.0
    return -np.inf

def log_probability(theta,cosmo):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + loglkl(theta,cosmo=cosmo)

def init_state():
    return np.random.uniform(0.2,0.4),np.random.uniform(0.01,0.1),np.random.uniform(60,80)
```

```python
import sys
sys.path.append('/Users/zhiyulu/Documents/Github/CourseCosmology2024/Sampler')
```

```python
from mcmc import Sampler
#log_probability(theta,m)

s=Sampler(loglkl=log_probability,
         init_state=init_state,
         step_size=[0.01,0.01,1],cosmo=m)

```

```python
chain=s.get_chain(100)

```
