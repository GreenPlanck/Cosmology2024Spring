# Estimate the power spectrum from map

It's interesting to see the effect of applying a mask

```python
cl_nonmask = hp.anafast(Tmap)
map_masked=Tmap*mask
cl_masked = hp.anafast(map_masked,use_pixel_weights=True)
sky_fraction = len(mask[mask!=0])/len(mask)
```

```python
ell=np.arange(np.shape(cl_masked)[0])
w_ell = hp.gauss_beam(5/60/180*np.pi, lmax=np.max(ell))
factor=ell*(ell+1)/2/np.pi
factor/=sky_fraction/w_ell  # some corrections
```

```python
plt.plot(ell,cl_nonmask*factor,'b',alpha=0.3)
plt.plot(ell,cl_masked*factor,alpha=0.3)
plt.xlim(2,1500)
plt.ylim(0,6000)
plt.xlabel('l')
plt.ylabel(r'$D_l[\mu k]^2$')
```

# Get the theory!

```python
import camb
#Set up a new set of parameters for CAMB
pars = camb.CAMBparams()
#This function sets up with one massive neutrino and helium set using BBN consistency
pars.set_cosmology(H0=67.5)
pars.set_for_lmax(1500, lens_potential_accuracy=0);
```

```python
results = camb.get_results(pars)
powers =results.get_cmb_power_spectra(pars, CMB_unit='muK',raw_cl=True)['total'][:,0]
```

```python
l=np.arange(np.shape(powers)[0])
lfactor=l*(l+1)/2/np.pi
```

```python
plt.plot(l,powers*lfactor)
plt.plot(ell,cl_masked*factor,alpha=0.3,label='beam corrected')
plt.plot(ell,cl_masked*factor*w_ell,alpha=0.3,label='non beam corrected')
plt.legend()

plt.xlabel('l')
plt.ylabel(r'$D_l[\mu k]^2$')
```
